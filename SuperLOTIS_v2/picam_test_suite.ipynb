```python
PrincetonInstruments.list_cameras()
```

```python

# SOPHIA camera SN
cam = PrincetonInstruments.PicamCamera('X030000923')

# Set exposure time to zero
cam.set_attribute_value("Exposure Time", 0) # ms

# Get all attribute values of the camera (dict that can be stored as FITS headers)
cam.get_all_attribute_values()

# To change temperature setpoint

# Get temperature only
cam.cav["Sensor Temperature Reading"]
cam.get_attribute_value("Sensor Temperature Reading")

# Set temperature
cam.set_attribute_value("Sensor Temperature Set Point", 10) # C

# Take an image
data = cam.grab(nframes=1, frame_timeout=10)


plt.figure()
im = plt.imshow(data[0], origin='lower')
plt.colorbar(im)
plt.show()
```
